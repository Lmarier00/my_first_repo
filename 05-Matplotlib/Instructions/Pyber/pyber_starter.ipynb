```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
city_data_df

# Display the data table for preview
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df
```

```python
# Combine the data into a single dataset
merged_table = pd.merge(city_data_df, ride_data_df, on="city")
```

```python
# Display the data table for preview
merged_table
```

## Bubble Plot of Ride Sharing Data

```python
# Breakdown Merged Table by City Type (Urban, Suburban, Rural) 
urban_type = merged_table[merged_table["type"] == "Urban"]
suburban_type = merged_table[merged_table["type"] == "Suburban"]
rural_type = merged_table[merged_table["type"] == "Rural"]
```

```python
# Calculate total rides, average fare and total Driver Count by Urban City Type
urban_ride_count = urban_type.groupby(["city"]).count()["ride_id"]
urban_avg_fare = urban_type.groupby(["city"]).mean()["fare"]
urban_driver_count = urban_type.groupby(["city"]).mean()["driver_count"]
urban_ride_count
```

```python
# Calculate total rides, average fare and total Driver Count by Suburban City Type
suburban_ride_count = suburban_type.groupby(["city"]).count()["ride_id"]
suburban_avg_fare = suburban_type.groupby(["city"]).mean()["fare"]
suburban_driver_count = suburban_type.groupby(["city"]).mean()["driver_count"]
```

```python
# Calculate total rides, average fare and total Driver Count by Rural City Type
rural_ride_count = rural_type.groupby(["city"]).count()["ride_id"]
rural_avg_fare = rural_type.groupby(["city"]).mean()["fare"]
rural_driver_count = rural_type.groupby(["city"]).mean()["driver_count"]
```

```python
# Add Title and labels for X and Y axis
plt.title("Pyber Ride Sharing Data (2018)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid(True)

plt.scatter(urban_ride_count,
           urban_avg_fare,
           s=10*urban_driver_count, c="lightcoral",
           edgecolor="black", linewidths=1, marker="o",
           alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
           suburban_avg_fare,
           s=10*suburban_driver_count, c="lightskyblue",
           edgecolor="black", linewidths=1, marker="o",
           alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
           rural_avg_fare,
           s=10*rural_driver_count, c="gold",
           edgecolor="black", linewidths=1, marker="o",
           alpha=0.8, label="Rural")
```

```python
# Create a legend
lgnd = plt.legend(fontsize="small", mode="Expanded",
                 numpoints=1, scatterpoints=1,
                 loc="best", title="City Types",
                 labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(34, 33, "Note:\nCircle size correlates\n   with driver\n      count per city.")
```

```python
# Save Figure
plt.savefig("../Images/PyberRideSharing.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Total Amount of Fares
total_city_fare = (merged_table["fare"]).sum()
total_city_fare
```

```python
# Breakdown Fares by City Type
fare_by_type = merged_table.groupby(["type"]).sum()["fare"]
fare_by_type
```

```python
# Calculate Type Percents
fare_spend_perc = (fare_by_type / total_city_fare) * 100
fare_spend_perc.round(2)
```

```python
# Build Pie Chart

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

explode = (0,0,0.1)
```

```python
# Show Figure
plt.title("% of Total Fares by City Type")
plt.pie(fare_spend_perc, explode=explode, labels=labels, colors=colors, autopct="%2.2f%%", shadow=True, startangle=140)
```

```python
# Save Figure

plt.savefig("../Images/FarebycityType.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Total Number of Rides
total_city_ride = (merged_table["ride_id"]).count()
total_city_ride
```

```python
# Total Number of Rides per city type
ride_by_type = merged_table.groupby(["type"]).count()["ride_id"]
ride_by_type
```

```python
ride_type_percent = (ride_by_type / total_city_ride) * 100
ride_type_percent.round(2)
```

```python
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]
```

```python
# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]
```

```python
explode = (0,0,0.1)
```

```python
plt.title("% of Total Rides by City Type")
plt.pie(ride_type_percent, explode=explode, labels=labels, colors=colors, autopct="%2.2f%%", shadow=True, startangle=140)
```

```python
# Save Figure
plt.savefig("../Images/RidesbycityType.png")
plt.show()
```

## Total Drivers by City Type


```python
driver_total = (city_data_df["driver_count"]).sum()
driver_total
```

```python
driver_type_totals = city_data_df.groupby(["type"]).sum()["driver_count"]
driver_type_totals
```

```python
driver_type_perc = (driver_type_totals / driver_total) * 100
driver_type_perc.round(2)
```

```python
# Build Pie Charts
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

explode = (0,0,0.1)
```

```python
plt.title("% of Total Drivers by City Type")

plt.pie(driver_type_perc, explode=explode, labels=labels, colors=colors, autopct="%2.2f%%", shadow=True, startangle=140)

plt.axis("equal")

# Save Figure
plt.savefig("../Images/DriversbyCityType.png")
plt.show()
```
